```python
import pandas as pd
```

```python
df = pd.read_csv('Meteo_corse_cleaned.csv', sep=';')
pd.set_option('display.max_columns', None)
df.head()
```

```python
# ajout de colonne sur les précispitation moyenne par an et mois
df['moyenne precipitations année'] = df.groupby('année')['RR'].transform('mean')
df['moyenne precipitations mois'] = df.groupby('mois')['RR'].transform('mean')
```

```python
# moyenne ecapotranspiration par mois et année
df['moyenne evapotranspiration année'] = df.groupby('année')['ETPMON'].transform('mean')
df['moyenne evapotranspiration mois'] = df.groupby('mois')['ETPMON'].transform('mean')
```

```python
# moyenne vitesse de vent par mois et année
df['moyenne vitesse vent année'] = df.groupby('année')['FFM'].transform('mean')
df['moyenne vitesse vent mois'] = df.groupby('mois')['FFM'].transform('mean')
```

```python
# moyenne température par mois et année
df['moyenne temperature année'] = df.groupby('année')['TM'].transform('mean')
df['moyenne temperature mois'] = df.groupby('mois')['TM'].transform('mean')
```

```python
df.head()
```

```python
df.to_csv('Meteo_corse_cleaned_features.csv', sep=';', index=False, encoding='utf-8')
```
